# Creating a Custom Ellipsoid

We can use the `ellipsoid_template` function in `constants.py` to get a list of the ellipsoid parameters that we need to fill in for a custom ellipsoid. Depending on the application, you only need to pass the parameters that are relevant.

````python
custom = constants.ellipsoid_template()

for key, value in custom.items():
    print(f"{key:20}: {value}")
````

Now, we can create a custom ellipsoid using `custom_ellipsoid` function

````python
my_custom_ellipsoid = constants.custom_ellipsoid(
    semi_major_axis=6378136.46,
    semi_minor_axis=6356755.0,
    GM=3.985e14,
    f=1/300.0
)

print("My Custom Ellipsoid Parameters:\n")
for key, value in my_custom_ellipsoid.items():
    print(f"{key:20}: {value}")
````

Notice that only those parameters we supplied are populated!

````python
constants.resolve_ellipsoid('wgs84')['GM']
````

If a required parameter is not passed to the custom ellipsoid, an error is raised!

````python
constants.require_ellipsoid_params(
    ellipsoid={'semi-minor': 800000.0, 'GM': 3.986e14, 'w': 7e-11},
    required=['semi-major', 'GM', 'J2']
)
````

You would use `constants.resolve_ellipsoid` to create a custom ellipsoid with the full parameter list currently implemented in `geoidlab`. However, those parameters that you do not supply are populated with `nan` values. If any required parameter is not supplied, an Exception is raised as in the previous code cell.

````python
custom = {
    'semi_major': 6378136.46,
    'semi_minor': 6356755.0,
    'GM': 3.985e14,
    'f': 1/300.0
}
ref_ellipsoid = constants.resolve_ellipsoid(ellipsoid=custom)
ref_ellipsoid
````

The `require_ellipsoid_params` function checks if the parameters required are supplied for the specific `context` and raises an Exception if not. We will see how this works with the previous example, but this time, we will provide a context.

````python
help(constants.require_ellipsoid_params)
````

````python
constants.require_ellipsoid_params(
    ellipsoid={'semi-minor': 800000.0, 'GM': 3.986e14, 'w': 7e-11},
    required=['semi-major', 'GM', 'J2'],
    context='Fame'
)
````

The ValueError message now specifically mentions the context, namely `Fame`. Wonderful!

If you want to interact with `geoidlab` via the config file, you may do the following:

- Create a local copy of geoidlab.cfg using `geoidlab -c`
- Open and edit the config file. Under `[topography]`, locate and update `ellipsoid` with your custom ellipsoid parameters.
- Execute by entering `geoidlab [subcommand] geoidlab.cfg`

Now we are going to remove `"semi_major":6378137.0` from our custom ellipsoid and see how `GeoidLab` handles this. 

````python
import subprocess

cmd = [
    "geoidlab", "ggm",
    "-m", "EGM2008",
    "-n", "360",
    "--do", "gravity-anomaly",
    "-b", "-180", "180", "-90", "90",
    "-gs", "1",
    "-gu", "degrees",
    "--ellipsoid", '{"e2":0.00669437999014,"GM":3.986004418e14,"C20":-4.84166774985e-4,"C40":7.90303733511e-7,"C60":-1.68724961151e-9,"C80":3.46052468394e-12,"C100":-2.65002225747e-15}',
    "-pn", "GeoidProject"
]

# Run with capture_output=True and text=True to get strings
result = subprocess.run(cmd, capture_output=True, text=True)

if result.returncode != 0:
    print("Command failed with error:")
    print(result.stderr)          # <-- This prints the actual error message
else:
    print("Success:", result.stdout)
````
